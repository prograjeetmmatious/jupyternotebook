```python
from sklearn.impute import KNNImputer
from sklearn.metrics import accuracy_score
import pandas as pd 
import numpy as np 
df=pd.read_csv('train (2).csv',usecols=['Pclass','Age','Fare','Survived'])
from sklearn.model_selection import train_test_split
x=df.drop(columns=['Survived'])
# y=df['Survived']  ## it returns a series 
y=df[['Survived']]  ## it return a data frame 
x_train,x_test,y_train,y_test=train_test_split(x,y)
im=KNNImputer(n_neighbors=5,weights='distance')
x_train_trf=im.fit_transform(x_train)
x_test_trf=im.fit_transform(x_test)
# remember x_train_trf is i array form and y is in dataframe avoid it
y_train=np.array(y_train)
y_test=np.array(y_test)
from sklearn.linear_model import LogisticRegression 
li=LogisticRegression ()
li.fit(x_train_trf,y_train)
y_pred=li.predict(x_test_trf)
acc=accuracy_score(y_test,y_pred)
print(f'score in knn impute {acc}')
```
