```python
import pandas as pd 
import seaborn as sns 
df=pd.read_csv('data_science_job.csv')
```

```python
df.shape
```

***we are going to perform cca***
***if data missing percent is less than 5 percent then  we  perform cca ***

```python
df.isnull().mean()*100
```

```python
df.isnull().mean()*100
```

```python
z=[]
for col in df.columns:
  if (df[col].isnull().mean()*100<5 and df[col].isnull().mean()*100>0):
   z.append(col)
```

```python
df[z]
```

```python
new_df=df[z].dropna()
```

```python
def draw(col) :
     sns.kdeplot(df[col])
     sns.kdeplot(new_df[col])
    
                            
    
```

```python
draw('experience')
```

```python
draw('enrolled_university')
```
